```python
import pandas as pd
import matplotlib.pyplot as plt

ar_train_df = pd.read_csv("Data/Arabic/csvTrainImages.csv", header=None)
ar_test_df = pd.read_csv("Data/Arabic/csvTestImages.csv", header=None)

ar_train_df_label = pd.read_csv("Data/Arabic/csvTrainLabel.csv", header=None)
ar_test_df_label = pd.read_csv("Data/Arabic/csvTestLabel.csv", header=None)

en_train_df = pd.read_csv("Data/English/emnist-balanced-train.csv", header=None)
en_test_df = pd.read_csv("Data/English/emnist-balanced-test.csv", header=None)
```

```python
ar_train_df
```

```python
en_train_df_label = en_train_df[0]
en_test_df_label = en_test_df[0]
```

```python
en_train_df = en_train_df.drop(en_train_df.columns[0], axis=1)
en_test_df = en_test_df.drop(en_test_df.columns[0], axis=1)
```

```python
ar_train_counts = ar_train_df_label[0].value_counts().sort_index()
ar_test_counts = ar_test_df_label[0].value_counts().sort_index()
```

```python
en_train_counts = en_train_df_label.value_counts().sort_index()
en_test_counts = en_test_df_label.value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# --- Plot 1: Arabic Train ---
ax[0, 0].bar(ar_train_counts.index, ar_train_counts.values)
ax[0, 0].set_title("Arabic Train Label Frequency")
ax[0, 0].set_xlabel("Label")
ax[0, 0].set_ylabel("Count")
ax[0, 0].set_xticks(range(0, max(ar_train_counts.index) + 1, 2))

# --- Plot 2: Arabic Test ---
ax[0, 1].bar(ar_test_counts.index, ar_test_counts.values)
ax[0, 1].set_title("Arabic Test Label Frequency")
ax[0, 1].set_xlabel("Label")
ax[0, 1].set_ylabel("Count")
ax[0, 1].set_xticks(range(0, max(ar_test_counts.index) + 1, 2))

# --- Plot 3: English Train ---
ax[1, 0].bar(en_train_counts.index, en_train_counts.values)
ax[1, 0].set_title("English Train Label Frequency")
ax[1, 0].set_xlabel("Label")
ax[1, 0].set_ylabel("Count")
ax[1, 0].set_xticks(range(0, max(en_train_counts.index) + 1, 2))

# --- Plot 4: English Test ---
ax[1, 1].bar(en_test_counts.index, en_test_counts.values)
ax[1, 1].set_title("English Test Label Frequency")
ax[1, 1].set_xlabel("Label")
ax[1, 1].set_ylabel("Count")
ax[1, 1].set_xticks(range(0, max(en_test_counts.index) + 1, 2))

plt.tight_layout()
plt.show()
```
